```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
system_name = 'kite2d'
data_dir = '../data/'
```

```python
run_name = 'prova'
```

```python
# READING THE PARAMETERS
params = dict()
f = open(data_dir + system_name + '/' + run_name + "/ev_params.txt", "r")
for line in f.readlines():
    params[line[:line.find('\t')]] = float(line[line.find('\t')+1:])
print(params)
env_params = dict()
f = open(data_dir + system_name + '/param.txt', "r")
for line in f.readlines():
    env_params[line[:line.find('\t')]] = float(line[line.find('\t')+1:])
print(env_params)
```

```python
f = open(data_dir + system_name + '/' + run_name + "/ev_states.txt", "r")
st_traj = []
st_labels = f.readline().split()
for line in f.readlines():
    st_traj.append(line.split())
st_traj = np.array(st_traj, dtype='float')
#st_traj = np.loadtxt(data_dir + system_name + '/' + run_name + "/ev_states.txt", dtype='float')
xs = np.linspace(0, params['ev_time'], len(st_traj))
```

```python
ind_to_plot = [0,1]

for i in ind_to_plot:
    plt.plot(xs, st_traj[:,i], label=st_labels[i])
    
plt.legend()
```
