```python
def read_best_policy(path):
    f = open(path, "r")
    b = []
    for line in f.readlines():
        policy_at_state = np.array([int(x) for x in line.split()])
        b.append(policy_at_state)
    return b
```

```python
best_p = read_best_policy(data_dir+'/best_policy.txt')
best_p = np.array(best_p)
best_q = ut.read_best_quality(data_dir+'/best_quality.txt')
best_q = np.array(best_q)
```

```python
np.amax(best_q)
```

```python
print(best_q[0])
```

```python
n_alphas = len(env_params['alphas'])
n_banks = len(env_params['banks'])
tensor = []

if system_name != "kite3d":
    n_betas = len(env_params['beta_bins'])-1

    for n in range(n_alphas):
        table = []
        for i in range(n_banks):
            line = []
            for j in range(n_betas):
                flag = 0
                for k in range(8): 
                    if best_p[n*(n_banks*n_betas)+(i*n_betas)+j][k] != 0:
                        line.append(k%3)
                        flag = 1
                if flag == 0:
                    line.append(8%3)
            table.append(line)
        tensor.append(table)
else:
    for n in range(n_alphas):
        line = []
        for i in range(n_banks):
            flag = 0
            for k in range(8): 
                if best_q[n*n_banks+i][k] != 0:
                    line.append(k)
                    flag = 1
            if flag == 0:
                line.append(8)
        tensor.append(line)

print(tensor)
                
        
```

```python
tensor_s = np.copy(tensor)
tensor_s = tensor_s.astype("str")
if system_name == "kite3d":
    tensor_s = np.transpose(tensor_s)
else:
    for n in range(n_alphas):
        tensor_s[n] = np.flip(tensor_s[n],0)
print(tensor_s)
```

```python
if system_name != "kite3d":
    for n in range(n_alphas):
        for i in range(n_banks):
            for j in range(n_betas):
                if tensor[n][i][j] == 0:
                    tensor_s[n][i][j] = ""
                else:
                    if tensor[n][i][j] == 1:
                        tensor_s[n][i][j] = ""
                    else:
                        tensor_s[n][i][j] = ""

if system_name == "kite3d":
    for n in range(n_alphas):
        for i in range(n_banks):
            if tensor[n][i] == 0:
                    tensor_s[i][n] = ""
            if tensor[n][i] == 1:
                tensor_s[i][n] = ""
            if tensor[n][i] == 2: 
                    tensor_s[i][n] = ""
            if tensor[n][i] == 3: 
                    tensor_s[i][n] = ""
            if tensor[n][i] == 4: 
                    tensor_s[i][n] = ""
            if tensor[n][i] == 5: 
                    tensor_s[i][n] = ""
            if tensor[n][i] == 6: 
                    tensor_s[i][n] = ""
            if tensor[n][i] == 7: 
                    tensor_s[i][n] = ""
            if tensor[n][i] == 8:
                    tensor_s[i][n] = ""
            
print(tensor_s)
```

### Quality trajectory for attack angles and vrel angles

```python
n_alphas=len(env_params['alphas'])
n_betas=len(env_params['beta_bins'])-1
alpha = n_alphas-3
betas = np.arange(n_betas)
colors = plt.cm.jet(np.linspace(0,1,len(betas)))

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(15,4))
states = range(0,len(q_traj[0]))

ax1.set_xlabel('Learning step', fontsize=14)
ax1.set_ylabel('Quality of decreasing angle', fontsize=14)
xs = np.linspace(0, alg_params['n_steps'] , alg_params['traj_points']+1 )
ax1.set_xlim([0,alg_params['n_steps']])
#ax1.set_ylim([-100,alg_params['init_values']])
count=0
for b in betas:
    index = b+alpha*n_betas
    ax1.plot(xs, q_traj[:,index,0], label=str(index) + " " + s_labels[index], c=colors[count])
    count+=1
#ax1.legend()

ax2.set_xlabel('Learning step', fontsize=14)
ax2.set_ylabel('Quality of staying', fontsize=14)
ax2.set_xlim([0,alg_params['n_steps']])
#ax1.set_ylim([-100,alg_params['init_values']])
count=0
for b in betas:
    index = b+alpha*n_betas
    ax2.plot(xs, q_traj[:,index,1], label=s_labels[index], c=colors[count])
    count+=1

ax3.set_xlabel('Learning step', fontsize=14)
ax3.set_ylabel('Quality of increasing', fontsize=14)
ax3.set_xlim([0,alg_params['n_steps']])
#ax1.set_ylim([-100,alg_params['init_values']])
count=0
for b in betas:
    index = b+alpha*n_betas
    ax3.plot(xs, q_traj[:,index,2], label=s_labels[index], c=colors[count])
    count+=1
    
plt.tight_layout()
plt.show()
```

### Quality trajectory for attack angles, bank and vrel angles

```python
fig, axs = plt.subplots(3,3, figsize=(15,10))
n_traj = len(q_traj[0])
colors = plt.cm.jet(np.linspace(0,1,n_traj))
xs = np.linspace(0, alg_params['n_steps'] , alg_params['traj_points']+1 )

for a_ind in range(3):
    for b_ind in range(3):
        pl_ind = b_ind + 3*a_ind
        axs[a_ind][b_ind].set_xlabel('Episode')
        axs[a_ind][b_ind].set_ylabel(a_labels[pl_ind])
        #axs[a_ind][b_ind].set_xlim([50000,100000])
        #axs[a_ind][b_ind].set_ylim([0,1500])
        count=0
        for index in range(n_traj):
            roll = 1
            if index%roll==0:
                axs[a_ind][b_ind].plot(i_traj[:,0], q_traj[:,count,pl_ind], label=s_labels[count], c=colors[count], lw=2)
                count += roll
#plt.legend()
plt.tight_layout()
plt.savefig(plot_dir+'qualities_eps.png', dpi=200)
plt.show()
```
