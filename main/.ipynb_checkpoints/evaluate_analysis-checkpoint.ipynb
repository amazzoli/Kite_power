```python
system_name = 'kite2d'
data_dir = 'data/'
```

```python
run_name = 'prova'

params = dict()

# EVALUATION PARAMETERS
params['n_steps'] = 10000
params['traj_steps'] = 10
```

```python
# WRITING THE PARAMETERS
f = open(data_dir + system_name + '/' + run_name + "/ev_params.txt", "a")
for k,v in params.items():
    f.write(k + "\t" + str(v) + "\n")
f.close()
```

```python
# COMPILING C++ SCRIPT
import subprocess
p = subprocess.Popen(["g++", "-o", "evaluate.exe", "evaluate.cpp", "lib/eval.cpp", "lib/utils.cpp", 
                      "lib/envs/kite.cpp", "lib/envs/"+system_name+".cpp", "-std=c++17"], 
stdout=subprocess.PIPE, stderr=subprocess.PIPE)
print(p.communicate()[1].decode('utf-8'))
#print(p.communicate()[1].decode('utf-8'))
```

```python
# EXECUTING THE SCRIPT
%time p = subprocess.Popen(["./evaluate.exe", system_name, run_name], stdout=subprocess.PIPE)
print(p.communicate()[0].decode('utf-8'))
```

```python
import numpy as np
v_traj = np.loadtxt(data_dir+run_name+'_v_traj.txt', dtype='float')
p_traj = np.loadtxt(data_dir+run_name+'_p_traj.txt', dtype='float')
r_traj = np.loadtxt(data_dir+run_name+'_r_traj.txt', dtype='float')
```

```python
def smooth_traj(traj, wind_size):
    new_traj, times = [], []
    i = 0
    while i < len(traj)-wind_size:
        new_traj.append(np.mean(traj[i:i+wind_size]))
        times.append(i + wind_size/2)
        i += wind_size
    return times, new_traj
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,4))
xs = np.arange(0, params['n_steps'], round(params['n_steps']/float(params['traj_steps'])))

ax1.set_xlabel('Time steps', fontsize=14)
ax1.set_ylabel('Probability of gathering', fontsize=14)
for k in range(0,params['max_k']+1):
    ax1.plot(xs, p_traj[:,k], label='k={}'.format(k))
    #print('k={}, final_gath_policy={}'.format(k, traj[-1]))
ax1.legend(fontsize=12)

ax2.set_xlabel('Time steps', fontsize=14)
ax2.set_ylabel('Value', fontsize=14)
for k in range(0,params['max_k']+1):
    ax2.plot(xs, v_traj[:,k], label='k={}'.format(k))
    #print('k={}, final_value={}'.format(k, v_traj[:,k][-1]))

ax2.legend(fontsize=12)

ax3.set_xlabel('N. of episodes', fontsize=14)
ax3.set_ylabel('Cumulative reward', fontsize=14)
wind_size=1000
times, new_traj = smooth_traj(r_traj, wind_size)
ax3.set_title('Smoothed with window of {} episodes'.format(wind_size))
ax3.plot(times,new_traj)


plt.tight_layout()
```
